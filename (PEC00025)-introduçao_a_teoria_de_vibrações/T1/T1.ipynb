Aplicou-se um deslocamento inicial e registrou-se a vibração pelo aplicativo iNVH.

```python
import numpy as np
import matplotlib.pyplot as plt
from   MRPy import MRPy
```

```python
data = MRPy.from_file('T1', form='invh')

data.printAttrib()
data.plot_time(fig=1);
```

```python
free = data.extract((1.9, 10), by='time')     

free.printAttrib()
free.plot_time(fig=2);
```

```python
afit, par = free.fit_decay()
t         = free.t_axis()               # prepare time axis for plotting

plt.figure(4, figsize=(18, 8), clear=True)
plt.plot(t, free[2])
plt.plot(t, afit[2], 'r:')
plt.legend(('original','fitted'))

plt.xlim(0, 8);  plt.xlabel('time (s)') 
plt.ylim(-15, 15);  plt.ylabel('amplitude (m)') 

plt.grid(True) 

print('Aplicando o código computacional, obteve-se os seguintes parâmetros:')
print('Amplitude:     {0:4.2f}m'.format(par[2,0]))
print('Frequency:     {0:4.2f}Hz  '.format(par[2,1]))
print('Damping ratio: {0:4.2f}%   '.format(par[2,2]*100))
print('Phase:         {0:4.2f}rad '.format(par[2,3]))
```

Observando os picos 8 e 27, respectivamente nos tempos aproximadamente de 1.9 e 5.2, estima-se que o amortecimento é 0.58%.
